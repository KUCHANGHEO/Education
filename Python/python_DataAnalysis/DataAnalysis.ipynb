## 4.5 선형대수

파이썬 에서는 두개의 2차원 배열을 * 연산자로 곱하면 행렬 곱셈이 아니라

대응하는 각각의 원소의 곱을 계산한다

행렬 곱셈은 배열 메소드이자 numpy 네임스페이스 안에 있는 dot함수를 이용해서 계산한다

```python
# inv 정사각 행렬의 역행렬을 계산한다
arr1 = np.array([[2,4],[1,3]])
np.linalg.inv(arr1)
```

```python
# det 행렬식을 계산한다
np.linalg.det(arr1)
```

```python
# 행렬의 대각선 원소의 합을 계산한다
np.trace(arr1)
```

```python
# A가 정사각 행렬일때 Ax = b 를 만족하는 x를 구한다
arr = np.array([[2,4],[1,8]])
arr1 = np.array([6,10])

np.linalg.solve(arr,arr1)
```

```python
# diag 정사각 행렬의 대각/비대각 원소를 1차원 배열로 반환하거나 1차원 배열을 대각선 원소로 하고 나머지 는 0으로 채운 단위행렬을 반환한다.
np.diag(np.array([1,2,3]))
```
