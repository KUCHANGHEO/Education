```python
import pandas as pd
```

```python
# Raw Data Loading
mnist = pd.read_csv('./mnist_train.csv')
```

```python
mnist.head()
```

```python
mnist.describe()
```

```python
y_data= mnist['label']
```

```python
y_data.head()
```

```python
X_data = mnist.drop('label', axis = 1, inplace=False)
```

```python
X_data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_data)
X_norm_data = scaler.transform(X_data)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_norm_data, y_data, test_size=0.2)
```

```python
X_train.shape
```

```python
# model 설정
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD
```

```python
model = Sequential()
```

```python
model.add(Flatten(input_shape=(784,))) # input Layer
model.add(Dense(units=10, activation= 'softmax'))
```

```python
model.compile(optimizer = SGD(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train, epochs=10, verbose=1, validation_split=0.2)
```

```python
model.evaluate(X_test , y_test)
```
